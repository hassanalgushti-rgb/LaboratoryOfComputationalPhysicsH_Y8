```python
%%bash
set -e

# =======================
# Ex 2 â€” One-cell (Linux)
# =======================

# ---- EDIT THESE ----
DATA_CSV="data.csv"   # path to your input file (e.g. ./data.csv or ~/Downloads/data.csv)
N=3                   # for 2D: how many copies to generate
# --------------------

# -----------------------
# 2A) Create data.txt:
#     - remove metadata (lines starting with #)
#     - replace commas with spaces
# -----------------------
grep -v '^#' "$DATA_CSV" | tr ',' ' ' > data.txt
echo "2A) Created data.txt"
head -n 5 data.txt
echo

# -----------------------
# 2B) Count even numbers
#     - extract all integers
#     - keep only even ones
#     - count them
# -----------------------
grep -oE '[0-9]+' data.txt | awk '$1 % 2 == 0' | wc -l | awk '{print "2B) Even numbers:", $1}'
echo

# -----------------------
# 2C) Compare sqrt(x^2+y^2+z^2) with 100*sqrt(3)/2
#     Count how many rows are Greater / Smaller
# -----------------------
awk '{
  x=$1; y=$2; z=$3;
  d=sqrt(x*x + y*y + z*z);
  thr=100*sqrt(3)/2;
  if (d > thr) high++; else low++;
}
END{
  print "2C) Greater:", high;
  print "2C) Smaller:", low;
}' data.txt
echo

# -----------------------
# 2D) Create N copies:
#     i-th file has all numbers divided by i (i=1..N)
# -----------------------
for i in $(seq 1 "$N"); do
  awk -v div="$i" '{
    for (j=1; j<=NF; j++) $j = $j/div;
    print
  }' data.txt > "data_divided_${i}.txt"
done
echo "2D) Created files: data_divided_1.txt ... data_divided_${N}.txt"
ls -lh data_divided_*.txt
```
